# Initialization

```python
import scqubits as qubit
import numpy as np
```

```python
N = 2  # number of big capacitors; changing to N=3 increases runtime but yields multiple inequivalent minima
ECB = 0.2
ECJ = 5.0
ECg = 10.0
flux = 0.0
EJ = 10.0
EJlist = np.array([EJ for j in range(2*N)])
nglist = np.array([0.0 for j in range(2*N-1)])
ncut = 8
param_dict = {"N":N, "ECB":ECB, "ECJ":ECJ, "ECg":ECg, 
              "EJlist":EJlist, "nglist":nglist, "flux":flux,
              "ncut":ncut, "truncated_dim":6}
CM = qubit.CurrentMirror(**param_dict)
```

# Eigensystem

```python
evals, evecs = CM.eigensys()
```
